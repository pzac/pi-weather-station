```python
import sqlite3

# import matplotlib
# matplotlib.use('nbagg')

from matplotlib import pyplot as plt

sqlite_file = 'data.db'
conn = sqlite3.connect(sqlite_file)
c = conn.cursor()

c.execute('SELECT * FROM data')
data = c.fetchall()

dates = []
ext_temp = []
int_temp = []
bar_temp = []
humidity = []
brightness = []
pressure = []
motion = []

for row in data:
    dates.append(row[1])
    ext_temp.append(row[2])
    brightness.append(row[3])
    int_temp.append(row[4])
    bar_temp.append(row[5])
    humidity.append(row[6])
    pressure.append(row[7])
    motion.append(row[8])
```

```python


plt.plot(dates, ext_temp, '-', label="Ext Temp", color='r')
plt.plot(dates, int_temp, '-', label="Int Temp", color='g')
plt.plot(dates, bar_temp, '-', label="Bar Temp", color='b')

plt.legend(loc = 'best')
plt.grid(True)
plt.ylabel("Temperature [C]")
plt.show()
```

```python
plt.plot_date(dates, ext_temp, '-', label="Ext Temp", color='r', xdate=True)
plt.plot_date(dates, int_temp, '-', label="Int Temp", color='g', xdate=True)
plt.plot_date(dates, bar_temp, '-', label="Bar Temp", color='b', xdate=True)

plt.legend(loc = 'best')
plt.grid(True)

ax = plt.axes() #(label="...")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.xticks(rotation=60)
plt.ylabel("Temperature [C]")
plt.show()
```

```python
plt.plot_date(dates, humidity, '-', label="Humidity", color='r', xdate=True)

plt.legend(loc = 'best')
plt.grid(True)

ax = plt.axes() #(label="...")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.xticks(rotation=60)
plt.ylabel("Humidity [%]")
plt.show()
```

```python
plt.plot_date(dates, pressure, '-', label="Pressure", color='r', xdate=True)

plt.legend(loc = 'best')
plt.grid(True)

ax = plt.axes() #(label="...")
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.xticks(rotation=60)
plt.ylabel("Pressure [hPa]")
plt.show()
```

```python
c.close()
conn.close()
```
